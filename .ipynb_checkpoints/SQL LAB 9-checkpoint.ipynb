```python
import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

import getpass
```

```python
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
data = pd.read_sql_query('SELECT * FROM actor', engine)
data.head()
```

```python
query = '''SELECT last_name, count(*) FROM actor
GROUP BY last_name
HAVING COUNT(*) = 1'''
data = pd.read_sql_query(query, engine)
data.head()
```

```python
query = '''SELECT *, weekday(rental_date) as day_of_rental, month(rental_date) as day_type, 
case
when weekday(rental_date)<3 then "workday"
else "weekend"
end as End_of_Week
from sakila.rental;'''
data = pd.read_sql_query(query, engine)
data.head()
```

```python
query = '''SELECT * FROM sakila.film
WHERE title = 'ARMAGEDDON';'''
data = pd.read_sql_query(query, engine)
data.head()

## this one might be incorrect because both didn't return armageddon.
```
